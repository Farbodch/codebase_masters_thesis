### Get error between the FEM and Integral-formulation

```python
import numpy as np
max_err_at_mesh_points_all = [[] for _ in range(mesh_interval+1)]
for exp_num_key in diffusion_1D_explicit.realizationDataDict.keys():
# for exp_num_key in range(1):
    for x_idx in range(mesh_interval+1):
        u_h_x_i = diffusion_1D.realizationDataDict[exp_num_key][x_idx]
        u_Integ_x_i = diffusion_1D_explicit.realizationDataDict[exp_num_key][x_idx]
        abs_diffs_at_x_i = np.abs(u_h_x_i-u_Integ_x_i)
        max_err_at_mesh_points_all[x_idx].append(abs_diffs_at_x_i)

max_err_at_mesh_points_max = []
for x_idx in range(len(max_err_at_mesh_points_all)):
    max_err_at_mesh_points_max.append(np.max(max_err_at_mesh_points_all[x_idx]))
```

#### Plot boxplot of the $|u_h - u_{intergal}|$ for all realizations

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
from os import path, makedirs

save_fig_toggle = True

save_directory = 'figs/1d_diffusion/realizations/error'
fig_name = f'1d_diffusion_boxPlot_err_explicitVfem_h_{mesh_interval}_numOfRealization_{num_of_model_realizations}.pdf'

base_fontsize = 20
rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'STIXGeneral'

h = len(max_err_at_mesh_points_all) - 1
positions = np.linspace(0.0, 1.0, h + 1)
spacing = positions[1] - positions[0]
width = 0.6 * spacing
fig, ax = plt.subplots(figsize=(18,8))

ax.boxplot(
    max_err_at_mesh_points_all,
    positions=positions,
    widths=width,
    patch_artist=True, 
    showfliers=True,
    showmeans=True,
    meanprops={"marker": "^",        
            "markerfacecolor": "red",   
            "markeredgecolor": "black",
            },
    medianprops={"color": "black"}
)

ax.set_xlim(-width, 1.0 + width)
nticks = 6  
tick_positions = np.linspace(0.0, 1.0, nticks)
ax.set_xticks(tick_positions)
ax.set_xticklabels([f"{p:.1f}" for p in tick_positions])
plt.xlabel("Spatial Domain", fontsize=base_fontsize+2)
plt.ylabel(r"$|u_h(x_i) - u(x_i)|$", fontsize=base_fontsize+2)
plt.xticks(fontsize=base_fontsize)
plt.yticks(fontsize=base_fontsize)

offset_text = plt.gca().yaxis.get_offset_text()
offset_text.set_fontsize(base_fontsize-2)
plt.minorticks_on()
plt.grid(True, which='both')
if save_fig_toggle:
    if not path.exists(save_directory):
        makedirs(save_directory)
    plt.savefig(f"{save_directory}/{fig_name}", dpi=900, bbox_inches='tight')
plt.show()
```

#### Plot scatter-plot of the $max_{x_i} |u_h(x_i) - u_{intergal}(x_i)|$, $i\in\{1,...,h+1\}$

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
from os import path, makedirs

save_fig_toggle = True

save_directory = 'figs/1d_diffusion/realizations/error'
fig_name = f'1d_diffusion_max_err_explicitVfem_h_{mesh_interval}_numOfRealization_{num_of_model_realizations}.pdf'

base_fontsize = 20
rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'STIXGeneral'

x = np.linspace(0,1,len(max_err_at_mesh_points_max))
fig = plt.figure(figsize=(18,8))

plt.scatter(x, max_err_at_mesh_points_max, color="firebrick", s=15)

plt.xlabel("Spatial Domain", fontsize=base_fontsize+2)
plt.ylabel(r"$\max_{x_i} |u_h(x_i) - u(x_i)|$", fontsize=base_fontsize+2)
# plt.minorticks_on()
plt.xticks(fontsize=base_fontsize)
plt.yticks(fontsize=base_fontsize)

offset_text = plt.gca().yaxis.get_offset_text()
offset_text.set_fontsize(base_fontsize-2)
plt.minorticks_on()
plt.grid(True, which='both')
if save_fig_toggle:
    if not path.exists(save_directory):
        makedirs(save_directory)
    plt.savefig(f"{save_directory}/{fig_name}", dpi=900, bbox_inches='tight')
plt.show()
```
